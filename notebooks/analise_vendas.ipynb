### Etapa 3: Tratamento de Valores Nulos

- Tratar valores ausentes em *[ produtos ]*
- Completar *[ categorias ]* com base no *[ produto ]*
- Recuperar *[ quantidade ]* usando *[ valor_total ]* e *[ preco ]*
- Recuperar *[ preco ]* usando *[ valor_total ]* e *[ preco ]*
- Calcular *[ valor_total ]* ausente
- Preencher *[ clientes ]* nao identificados
- Preencher *[ estados ]* nao informados

```python
df
```

```python
df["produto"] = df["produto"].fillna("Produto_Nao_Informado")
df
```

- Produtos antes NaN foram substituidos pela generalizacao "Produto_Nao_Informado", visando nao perder as informacoes de vendas 

```python
df["categoria"].unique()
```

```python
df["produto"].unique()
```

```python
# Imputando moveis sem categoria
df.loc[df["produto"].isin(["Cadeira Gamer", "Mesa Escritório"]), "categoria"] = "Móveis"
df.info()
```

```python
df.loc[df["produto"].isin(["Livro Banco de Dados", "Livro Python"]), "categoria"] = "Livros"
df.info()
```

```python
df.loc[df["produto"].isin(["Teclado Mecânico", "Smartphone", "Notebook", "Fone Bluetooth"]), "categoria"] = "Eletrônicos"
df.info()
```

```python
df.loc[df["produto"].isin(["Curso Online"]), "categoria"] = "Educação"
df.info()
```

```python
df.loc[df["produto"].isin(["Mochila"]), "categoria"] = "Acessórios"
df.info()
```

- Campos de categoria recuperados a partir do produto

```python
df
```

```python
df["quantidade"] = df["quantidade"].fillna(df["valor_total"] / df["preco_unitario"])
df.info()
```

```python
df[df["quantidade"].isnull()]
```

- Alguns campos de *[ preco_unitario ]* estao NaN, mas temos a informacao do preco unitario de cada produto

```python
df["produto"].unique()
```

```python
precos = {
    "Cadeira Gamer": 1200,
    "Teclado Mecânico": 350,
    "Smartphone": 2200,
    "Notebook": 3500,
    "Mesa Escritório": 800,
    "Fone Bluetooth": 250,
    "Curso Online": 500,
    "Livro Banco de Dados": 120,
    "Livro Python": 90,
    "Mochila": 180
}
```

```python
df["preco_unitario"] = df["preco_unitario"].fillna(
    df["produto"].map(precos)
)
```

```python
df.info()
```

- Campos de *[ preco_unitario ]* tratado, valores NaN substituidos pelos valores corretos

```python
df["quantidade"] = df["quantidade"].fillna(df["valor_total"] / df["preco_unitario"])
df.info()
```

- Finalizacao da task: Recuperar Quantidade usando Valor_Total e Preço

```python
df
```

```python
df["valor_total"] = df["valor_total"].fillna(df["preco_unitario"] * df["quantidade"])
df.info()
```

- campos NaN na coluna *[ valor_total ]* preenchidos a partir do calculo: *df["preco_unitario"]* x *df["quantidade"]*

```python
df.info()
```

```python
df["cliente"] = df["cliente"].fillna("cliente_nao_informado")
df.info()
```

- Campos NaN na coluna *[ cliente ]* preenchidos com "cliente_nao_informado"

```python
df.info()
```

```python
df["estado"] = df["estado"].fillna("ED")
df.info()
```

- Campos NaN na coluna *[ estado ]* preenchidos com "ED" (Estado Desconhecido)
